```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression


from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x_train=d1.drop("y",axis=1)
```

```python
x_train
```

```python
y_train=d1['y']
```

```python
y_train
```

```python
d2=pd.read_csv("bank.csv")
```

```python
x_test=d2.drop('y',axis=1)
y_test=d2['y']
```

```python
encoding_columns = list(x_train.select_dtypes(include=object))
encoding_columns
```

```python
x_train['education'].unique()
```

```python
values = {'unknown': 0, 'primary': 1,'secondary': 2,'tertiary': 3}

x_train['education'] = x_train['education'].replace(values)
x_test['education'] = x_test['education'].replace(values)
```

```python
x_train['education'].value_counts()
```

```python
cols = list(x_train.select_dtypes(include =object)) 
x_train = pd.get_dummies(x_train,columns = cols,prefix = cols)
x_test = pd.get_dummies(x_test,columns=cols,prefix = cols)

x_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=0)
```

```python
model.fit(x_train,y_train)
```

```python
model_predict = model.predict(x_train)
```

```python
acc = accuracy_score(y_train,model_predict)
print(f"Accuracy Score:{acc}")
print("\n")
report = classification_report(y_train,model_predict)
print(f"Classification Report:{report}")
```

```python
y_model_predict = model.predict(x_test)
```

```python

y_acc = accuracy_score(y_test, y_model_predict)
print(f"Accuracy is:{y_acc}")

print("\n")

y_report = classification_report(y_test,y_model_predict)
print(f"Classification Report:{y_report}")
```
